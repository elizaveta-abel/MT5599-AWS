```python
# the data that currently is in Equinor/data/input_data/argentina_daily_tweets.csv is df
# this was done on 27.09.2022 for 1 day, and yielded 300k+ tweets

"""Twitter Scraping Tool"""
import json
import snscrape.modules.twitter as sntwitter
from _datetime import datetime, timedelta
from multiprocessing import Pool
import tqdm # for progress bar
import pytz # for current time
import pandas as pd
import itertools

#now = datetime.now().replace(tzinfo=pytz.utc)
#start_date = (now - timedelta(days=120)).strftime('%Y-%m-%d')
#end_date = (now + timedelta(days=1)).strftime('%Y-%m-%d')
#date_window = str(start_date) + "-" + str(end_date)
#path = "scaper_output.json"
scraper = sntwitter.TwitterSearchScraper

dates = pd.date_range(start="2015-11-01",end="2015-11-03").tolist()

def get_tweetlist(date):
    
    tweetlist = []
    start_date = date.strftime('%Y-%m-%d')
    end_date = (date + timedelta(days=1)).strftime('%Y-%m-%d')
    
    for idx, tweet in enumerate(scraper(f'since:{start_date} until:{end_date}').get_items()):

        tweetlist.append({'cashtags': str(tweet.cashtags),
        'content': str(tweet.content),
        'conversationId': str(tweet.conversatoinId),
        'coordinates': str(tweet.coordinates),
        'date': str(tweet.date),
        'hashtags': str(tweet.hashtags),
        'id': str(tweet.id),
        'inReplyToTweetId': str(tweet.inReplyToTweetId),
        'inReplyToUser': str(tweet.inReplyToUser),
        'json': str(tweet.json),
        'lang': str(tweet.lang),
        'likeCount': str(tweet.likeCount),
        'media': str(tweet.media),
        'mentionedUsers': str(tweet.mentionedUsers),
        'outlinks': str(tweet.outlinks),
        'outlinksss': str(tweet.outlinkss),
        'place': str(tweet.place),
        'quoteCount': str(tweet.quoteCount),
        'quotedTweet': str(tweet.quotedTweet),
        'renderedContent': str(tweet.renderedContent),
        'replyCount': str(tweet.replyCount),
        'retweetCount': str(tweet.retweetCount),
        'retweetedTweet': str(tweet.retweetedTweet),
        'source': str(tweet.source),
        'sourceLabel': str(tweet.sourceLabel),
        'sourceUrl': str(tweet.sourceUrl),
        'tcooutlinks': str(tweet.tcooutlinks),
        'tcooutlinksss': str(tweet.tcooutlinkss),
        'url': str(tweet.url),
        'user_created': str(tweet.user.created),
        'user_description': str(tweet.user.description),
        'user_descriptionUrls': str(tweet.user.descriptionUrls),
        'user_displayname': str(tweet.user.displayname),
        'user_favouritesCount': str(tweet.user.favouritesCount),
        'user_followersCount': str(tweet.user.followersCount),
        'user_friendsCount': str(tweet.user.friendsCount),
        'user_id': str(tweet.user.id),
        'user_json': str(tweet.user.json),
        'user_label': str(tweet.user.label),
        'user_linkTcourl': str(tweet.user.linkTcourl),
        'user_location': str(tweet.user.location),
        'user_mediaCount': str(tweet.user.mediaCount),
        'user_profileBannerUrl': str(tweet.user.profileBannerUrl),
        'user_profileImageUrl': str(tweet.user.profileImageUrl),
        'user_protected': str(tweet.user.protected),
        'user_rawDescription': str(tweet.user.rawDescription),
        'user_statusesCount': str(tweet.user.statusesCount),
        'user_url': str(tweet.user.url),
        'user_username': str(tweet.user.username),
        'user_verified': str(tweet.user.verified)})
        
    print(f"{start_date} completed!")
    return tweetlist


#  with open('../data/argentina_users.txt') as f:
#      handles = f.read().splitlines()
#  handles = ["MateoReyesLara", "LolaLo1974"]

pool = Pool(processes=len(dates))
results = []
for result in tqdm.tqdm(pool.imap_unordered(get_tweetlist, dates), total=len(dates)):
    results.extend(result)

#  for date in dates:
#      results = get_tweetlist(date)
#      with open("../data/all_argentina" + date.strftime('%Y-%m-%d') + ".json", "w", encoding="utf-8") as f:
#          json.dump(results, f, indent=4, ensure_ascii=False)

with open("../data/argentina_november_1.json", "w", encoding="utf-8") as f:
    json.dump(results, f, indent=4, ensure_ascii=False)
```
