# Twitter Worldwide

There are about 200 billion tweets posted yearly. 
8.46% of tweets (of 100 million sample) were in Spanish in 2018 (https://www.vicinitas.io/blog/twitter-social-media-strategy-2018-research-100-million-tweets#language).
So there are around 16 billion tweets posted in Spanish per year.
Only around 1% of tweets are geotagged, leaving a dataset of size around 160 million spanish geotagged tweets per year.
In 2021, 473.462.951 people spoke Spanish worldwide total, 45.808.747 of which were in Argentina
(https://cvc.cervantes.es/lengua/espanol_lengua_viva/pdf/espanol_lengua_viva_2021.pdf). This is 9.675% of the Spanish speaking population living in Argentina.
Realistically, we can expect ~15,480,407 tweets to be geotagged with Argentina per year.

If we get all the usernames from that dataset, we will get all Argentinian visitors.
Then we can scrape tweets from all these usernames.
Then label their residence status depending on where they tweet most frequently from.

```python
import pandas as pd

dates1 = pd.date_range(start='2015-12-01', end='2015-12-02', freq='H', closed='left')
dates2 = pd.date_range(start='2016-04-15', end='2016-04-16', freq='H', closed='left')
dates3 = pd.date_range(start='2016-10-30', end='2016-10-31', freq='H', closed='left')

dates = dates1.union(dates2).union(dates3)

datess = []
for i in range(len(dates)):
    datess.append(str(dates[i]).replace(" ", "_") + "_UTC")
dates = datess
```

```python
import boto3
import pandas as pd
from sagemaker import get_execution_role
import tqdm

role = get_execution_role()

files = []
DataFrameName = []
for date in dates:
    files.append("../../data/tweets_spanish_" + date + ".json")
    DataFrameName.append("tweets_spanish_" + date)

dfs = {}

for name, file in zip(DataFrameName, files):
    print(name, file)
    dfs[name] = pd.read_json(file)
    
df = pd.concat(dfs.values(), ignore_index=True)
df
```
