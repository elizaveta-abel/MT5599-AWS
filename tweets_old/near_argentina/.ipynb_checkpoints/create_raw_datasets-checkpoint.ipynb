# Objective: Create Raw Dataset of Tweets of Argentinian Residents and Visitors

So far, we collected all tweets from Argentina during specified time periods for 2016 and 2020 outbreaks.
Then used that dataset to extract residents and visitors. A second Twitter scraper procedure was run to collect all tweets from those residents and visitors during specified time periods.

In this notebook, these datasets were combined for further processing.

```python
import boto3
import pandas as pd
from sagemaker import get_execution_role

for i in range(0,253):
    obj = "spanish_tweets_2016_" + str(i) + ".json"
    obj_loc = "../../data/" + obj
    boto3.Session().resource('s3').Bucket("mt5599").Object("tweets/" + obj).upload_file(obj_loc)
```

```python
import boto3
import pandas as pd
from sagemaker import get_execution_role

role = get_execution_role()

files = []
DataFrameName = []
for i in range(28):
    files.append("s3://mt5599/raw_tweets/2016/tweets_place_2016_" + str(i) + ".json")
    DataFrameName.append("tweets_place_2016_" + str(i))

dfs = {}

for name, file in zip(DataFrameName, files):
    dfs[name] = pd.read_json(file)
    print(name, " done")
```

```python
df = pd.concat(dfs.values(), ignore_index=True)
df
```

```python
df["resident/visitor"] = "visitor"
```

```python
df_location = pd.read_json("../data/tweets_location_2019.json")
df_location["resident/visitor"] = "resident"
df_location
```

```python
df = pd.concat([df,df_location], ignore_index=True)
df
```

```python
df.DateTime.min()
```

```python
df.DateTime.max()
```

```python
import feather

df.to_feather("../data/df_2016.feather")
boto3.Session().resource('s3').Bucket("mt5599").Object("df_2016.feather").upload_file("../data/df_2016.feather")
```
