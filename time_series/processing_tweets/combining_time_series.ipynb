```python
import pandas as pd

import boto3
import pandas as pd
from sagemaker import get_execution_role
import tqdm

boto3.Session().resource('s3').Bucket("mt5599").Object("tweets/handles_combined.feather").upload_file("../../data/handles_combined.feather")
```

```python
# reading in cleaned files for combining into one dataframe

dfs = []

missed_files = []

for i in tqdm.tqdm(range(0, 568)):
    try:
        df = pd.read_feather("../../data/spanish_tweets_2016_" + str(i) + "_processed.feather")
        dfs.append(df)
    except:
        missed_files.append(i)
        continue
```

```python
# combining cleaned files into one dataframe

main_df = pd.concat(dfs, axis=0)
main_df
```

```python
main_df.drop_duplicates(inplace=True)
main_df.reset_index(drop=True, inplace=True)
main_df.reset_index(inplace=True)
```
