```python
val_count0 = df_0.place_country.value_counts(dropna=False).rename_axis('unique_values').reset_index(name='counts')
val_count1 = df_1.place_country.value_counts(dropna=False).rename_axis('unique_values').reset_index(name='counts')
val_count2 = df_2.place_country.value_counts(dropna=False).rename_axis('unique_values').reset_index(name='counts')
val_count3 = df_3.place_country.value_counts(dropna=False).rename_axis('unique_values').reset_index(name='counts')
```

```python
val_count = pd.concat([val_count0, val_count1, val_count2, val_count3])
```

```python
val_count0
```

```python
val_count
```

```python
#val_count.loc[0,"unique_values"] = "NaN"
val_count.loc[0,"unique_values"]
```

```python
val_count = val_count.groupby("unique_values", dropna=False).sum()
val_count
```

```python
val_count.sum()
```

```python
#141442344 +84,040,056
```

```python
val_count["% geotagged"] = val_count.counts / 84040056 * 100
val_count["% total"] = val_count.counts/225482400*100
val_count = val_count.sort_values(by="counts", ascending=False)
```

```python
val_count = val_count.round({"% geotagged":3, "% total":3})
val_count
```

```python
val_count = val_count.reset_index(drop=True)
val_count.to_csv("value_counts.csv", index=False)
```
